```python
#find lat and long of city center then subtract out lat and long form it to find the difference 
# weak correlation to test in model Floor waterfront view and grade 
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import numpy as np
from datetime import datetime
import statsmodels.formula.api as smf
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from scipy import stats
import math
from PIL import Image

df2 = pd.read_csv("kc_house_data_clean.csv")
```

```python
df2 =df2.drop(["Unnamed: 0"], axis = 1)
df2
```

```python
y = df2["price"]
X = df2.drop(columns=["price","Renovated"], axis=1)
```

```python
model = sm.OLS(y, sm.add_constant(X.astype(float)), missing = "drop").fit()
results = model.summary()
print(results)
```

```python
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score

linreg = LinearRegression()
cv_5_results  = np.mean(cross_val_score(linreg, X, y, cv=5,  scoring='r2'))
cv_5_results
```

```python
cv_5_results_rmse  = np.mean(cross_val_score(linreg, X, y, cv=5,  scoring='neg_root_mean_squared_error'))
cv_5_results_rmse
```

```python
cv_10_results = np.mean(cross_val_score(linreg, X, y, cv=10,  scoring='r2'))
cv_10_results
```

```python
cv_10_results_rmse = np.mean(cross_val_score(linreg, X, y, cv=10,  scoring='neg_root_mean_squared_error'))
cv_10_results_rmse
```
