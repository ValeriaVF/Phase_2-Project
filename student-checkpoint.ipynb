```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime
import statsmodels.formula.api as sm
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from scipy import stats
```

```python
#Read dataset
df = pd.read_csv("data/kc_house_data.csv")
```

```python
#Converting dates to quarters 
df['date'] = pd.to_datetime(df['date'])
df["quarter"] = df['date'].dt.quarter
```

```python
#Revist if we should drop all NAs
df = df.dropna()
```

```python
df.shape
```

```python
#Function to convert categorical data 
def convert_categorical(column_names):
    for column_name in column_names:
        df[column_name] = pd.Categorical(df[column_name])
        df[column_name] = df[column_name].cat.codes
```

```python
convert_categorical(["view","waterfront","condition","grade","date"])
```

```python
df.dtypes
```

```python
#Remove outliers
drop_idx = []
for columns in df:
    if df[columns].dtype not in (float, int):
        continue
    drop_idx.extend(df[np.abs(stats.zscore(df[columns])) >= 3].index)
df = df.drop(set(drop_idx))
```

```python
#Checking correlation between variables
#Highest correlation to price is sqft_living, bathrooms, 
df.corr().sort_values(by="price", ascending=False)
```

```python
#Dropping data with weak correlation
#Dropping sqft_living15 and sqft_above due to correlation to sqft_living,
df = df.drop(["lat", "long","sqft_basement","sqft_living15","sqft_above","sqft_lot15","id","sqft_lot","zipcode","yr_built","condition","date","quarter","yr_renovated"], axis=1)
```
